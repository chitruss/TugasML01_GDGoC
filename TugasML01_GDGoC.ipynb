```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(data=df, x="Accident Severity", palette="coolwarm")
plt.title("Distribusi Keparahan Kecelakaan")
plt.xticks(rotation=45)
plt.show()
```

```python
df["Urban/Rural"].value_counts().plot.pie(autopct="%1.1f%%", colors=["skyblue", "salmon"])
plt.title("Perbandingan Kecelakaan di Area Urban vs Rural")
plt.ylabel("")
plt.show()
```

```python
sns.scatterplot(data=df, x="Speed Limit", y="Number of Injuries", alpha=0.6)
plt.title("Hubungan antara Batas Kecepatan dan Jumlah Cedera")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="Weather Conditions", order=df["Weather Conditions"].value_counts().index)
plt.xticks(rotation=45)
plt.title("Jumlah Kecelakaan Berdasarkan Kondisi Cuaca")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Number of Vehicles Involved"], bins=10, kde=True, color="blue")
plt.title("Distribusi Jumlah Kendaraan yang Terlibat dalam Kecelakaan")
plt.xlabel("Jumlah Kendaraan")
plt.ylabel("Frekuensi")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="Number of Fatalities", y="Emergency Response Time")
plt.title("Hubungan Waktu Respons Darurat dengan Fatalitas")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Economic Loss"], bins=20, kde=True, color="red")
plt.title("Distribusi Kerugian Ekonomi akibat Kecelakaan")
plt.xlabel("Kerugian Ekonomi")
plt.ylabel("Frekuensi")
plt.show()
```

**Insight yang di dapatkan setelah melakukana visualisasi data :**

- Distribusi Keparahan Kecelakaan (Bar Plot) : Keparahan kecelakaan berdasarkan bar plot menunjukkan bahwa kecelakaan rata-rata, ringan, maupun berat berada di distribusi yang sama besarnya. Aksi yang bisa dilakukan adalah dengan meningkatkan penerapan rambu keselamatan di lokasi rawan kecelakaan, dan dapat meningkatkan kesadaran pengemudi dengan kampanye keselamatan berkendara.  

- Perbandingan Kecelakaan di Area Perkotaan vs Pedesaan (Pie Chart) : Berdasarkan hasilnya berbeda tipis yang terjadi banyak pada daerah pedesaan, kemungkinan besar terkait dengan jalan rusak, penerangan yang kurang, atau keterbatasan fasilitas medis di daerah tersebut. Aksi yang bisa dilakukan adalah dengan memperbaiki jalan dan penerangan, serta meningkatkan akses ambulans.  

- Hubungan Kecepatan vs Jumlah Cedera (Scatter Plot) : Sepertinya keduanya berhubungan dan juga mungkin terdapat beberapa faktor lain contohnya cuaca, kondisi jalan, atau faktor manusia yang lebih berpengaruh dibandingkan kecepatan. Aksi yang bisa dilakukan adalah menambah speed bump atau kamera tilang elektronik di daerah rawan kecelakaan dan meningkatkan pengawasan terhadap kendaraan yang melaju di atas batas kecepatan.  

- Pengaruh Cuaca terhadap Kecelakaan (Bar Plot) : Pada kondisi cuaca apapun tidak dapat menjadikan alasan valid terjadinya kecelakaan, karena di cuaca apapun pada data jumlah datanya tetap sama. Aksi yang bisa dilakukan adalah menambah rambu peringatan saat cuaca buruk dan memperbaiki drainase jalan untuk menghindari banjir, juga dapat meningkatkan edukasi bagi pengemudi tentang cara berkendara di cuaca ekstrem.  

- Jumlah Kendaraan Terlibat dalam Kecelakaan (Histogram) : Distribusi Jumlah Kendaraan yang Terlibat dalam Kecelakaan tidak hanya satu melainkan dapat lebih kendaraan. Aksi yang bisa dilakukan adalah menambah fasilitas rest area bagi pengemudi yang menempuh perjalanan jauh dan meningkatkan pengawasan terhadap pelanggaran lalu lintas seperti menerobos lampu merah.  

- Emergency Response Time vs Fatalitas (Box Plot) : Jika semakin lama waktu respons darurat = semakin tinggi jumlah korban, ini menunjukkan bahwa keterlambatan ambulans atau layanan darurat berkontribusi pada tingkat fatalitas kecelakaan, dan jika waktu respons tidak terlalu berpengaruh, mungkin faktor lain seperti kecepatan tabrakan atau kondisi korban sebelum kejadian lebih dominan. Aksi yang bisa dilakukan adalah meningkatkan jumlah pos ambulans di daerah rawan kecelakaan dan menambahkan jalur khusus untuk ambulans di jalan utama.  

- Kerugian Ekonomi akibat Kecelakaan (Histogram) : Visualisasi data menunjukkan bahwasanya terdapat kerugian ekonomi tinggi pada kecelakaan, ini menunjukkan bahwa kecelakaan berdampak signifikan terhadap biaya medis, perbaikan kendaraan, dan klaim asuransi. Aksi yang bisa dilakukan adalah mendorong program asuransi bagi pengemudi untuk mengurangi dampak ekonomi, dan meningkatkan peraturan terkait keselamatan kerja bagi pengemudi kendaraan komersial.  

Dari berbagai visualisasi, ada beberapa faktor utama yang tampaknya berperan dalam kecelakaan:  

1. Kecepatan tinggi meningkatkan jumlah korban kecelakaan.  
2. Cuaca buruk berkontribusi terhadap kecelakaan, terutama di jalan dengan visibilitas rendah.  
3. Keterlambatan layanan darurat berpengaruh terhadap tingkat fatalitas.  
4. Kecelakaan lebih banyak terjadi di perkotaan dibandingkan pedesaan, mungkin karena kepadatan lalu lintas.  
5. Faktor manusia seperti kelelahan, alkohol, dan perilaku berkendara juga memengaruhi tingkat kecelakaan.  

Untuk mengurangi risiko kecelakaan, perlu dilakukan berbagai tindakan seperti meningkatkan pengawasan lalu lintas, memperbaiki infrastruktur jalan, serta mempercepat layanan darurat.
